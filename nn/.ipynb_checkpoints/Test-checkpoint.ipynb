```python
# Celda 1: Verificar instalaciones

import sys

print(f"Python: {sys.version}")
print(f"TensorFlow: {tf.__version__}")
print(f"PyTorch: {torch.__version__}")

# Celda 2: Probar TensorFlow
print("üîç TensorFlow Info:")
print(f"Dispositivos disponibles: {tf.config.list_physical_devices()}")
print(f"GPU disponible: {len(tf.config.list_physical_devices('GPU')) > 0}")

# Celda 3: Probar PyTorch
print("\nüîç PyTorch Info:")
print(f"CUDA disponible: {torch.cuda.is_available()}")
if torch.cuda.is_available():
    print(f"Dispositivo CUDA: {torch.cuda.get_device_name()}")

# Celda 4: Ejemplo simple de TensorFlow
import tensorflow as tf

model = tf.keras.Sequential([
    tf.keras.layers.Dense(10, activation='relu', input_shape=(5,)),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy')
print("\n‚úÖ Modelo de TensorFlow creado exitosamente!")

# Celda 5: Ejemplo simple de PyTorch
import torch
import torch.nn as nn

class SimpleModel(nn.Module):
    def __init__(self):
        super(SimpleModel, self).__init__()
        self.linear1 = nn.Linear(5, 10)
        self.linear2 = nn.Linear(10, 1)
    
    def forward(self, x):
        x = torch.relu(self.linear1(x))
        x = torch.sigmoid(self.linear2(x))
        return x

model_pytorch = SimpleModel()
print("‚úÖ Modelo de PyTorch creado exitosamente!")
```
